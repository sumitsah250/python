```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,6,3,16,3,4,5,0]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['math',"c++","c","python"]
y1=[95,88,45,68]
plt.plot(x,y1,label='sumit')
y2=[23,44,55,76]
plt.plot(x,y2,label="bidur")
y3=[67,89,76,88]
plt.plot(x,y3,label="sujan")
plt.title("Marks Graph")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()  # this is used to show the color of line in seprate box
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7,8]
y=[2,6,3,16,3,4,5,0]
plt.scatter(x,y,label="stars",color='green',marker="1",s=30) # s is size of the point
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
items =['android','ios','windows','linux']
properties=[20,40,30,10]
colors = ['y','r','g','b']
plt.pie(properties,labels=items,colors=colors,startangle=20,shadow=True,explode=(0,0,0.1,0),radius=1.6,autopct='%5.6f%%')
plt.title('smart phones')
plt.legend(loc=(-0.5,-0.3))
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,6,3,16,3,4,5,0]
tick_label=["sumit","nabin","sumit","nabin","sumit","nabin","sumit","nabin"]
plt.bar(x,y,tick_label=tick_label,width=0.9,color=['r','k','g','b','r','k','g','y',])
plt.show()
```

```python
import matplotlib.pyplot as plt
ages=[2,4,66,55,44,77,88,23,43,22,23,25,64,54,33,21,21,22,3,22,32,33]
range =(0,50)
bins=10
plt.hist(ages,bins,range,color='blue',histtype='bar',rwidth=0.5)
plt.xlabel("ages")
plt.ylabel("No .of people")
plt.title("ages of people")
plt.show()
```
